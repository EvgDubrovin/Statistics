### ОДНОФАКТОРНЫЙ ДИСПЕРСИОННЫЙ АНАЛИЗ

В эксперименте сравнивается эффективность четырех различных типов терапии. 
Мы хотим понять, есть ли значимая взаимосвязь между типом терапии и показателем уровня экспрессии гена.

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\BSNS\STUDY\Stepik\Основы Статистики\genetherapy.csv')
```

```python
A = data[data['Therapy']=='A']['expr']
A
```

```python
B = data[data['Therapy']=='B']['expr']
```

```python
C = data[data['Therapy']=='C']['expr']
```

```python
D = data[data['Therapy']=='D']['expr']
```

```python
# Применяем однофакторный дисперсионный анализ к сериям наших данных
stats.f_oneway(A, B, C, D)
```

```python
# Получили: F = 8.037; p-value = 0.00015;
```

```python
# Построим график и интерпретируем результат
ax = sns.boxplot(x="Therapy", y="expr", data=data)
sns.set(rc={'figure.figsize':(9, 6)}) # размер графика

plt.title('Уровень экспрессии гена при различной терапии', fontsize=14)
plt.xlabel('Терапия', fontsize=14)
plt.ylabel('Уровень экпрессии гена', fontsize=14)
sns.despine()
```

```python
# Удалось выявить статистически значимую взаимосвязь между типом терапии и показателем уровня экпрессии гена (F = 8.04, p < 0.05)
```
